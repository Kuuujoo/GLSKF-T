```python
import numpy as np
from PIL import Image
import os

# 设置参数
seedr = 920
missing_rate = 0.9  # 缺失率90%
image_names = ['Airplane', 'House256', 'House512', 'Peppers', 'Tree', 'Sailboat', 'Female']

# 设置随机种子
np.random.seed(seedr)

# 创建输出目录
output_dir = './masked_images'
os.makedirs(output_dir, exist_ok=True)

# 处理每张图片
for img_name in image_names:
    img_path = f'./data/{img_name}.tiff'
    if not os.path.exists(img_path):
        continue
    
    # 读取图像
    I = np.array(Image.open(img_path))
    
    # 生成随机掩码并应用
    h, w = I.shape[:2]
    mask = np.random.rand(h, w) > missing_rate
    
    if len(I.shape) == 3:  # 彩色图像
        masked_image = I.copy()
        for c in range(I.shape[2]):
            masked_image[:, :, c] = I[:, :, c] * mask
    else:  # 灰度图像
        masked_image = I * mask
    
    # 保存添加掩码后的图像
    Image.fromarray(masked_image.astype(I.dtype)).save(f'{output_dir}/{img_name}_masked.png')
```
